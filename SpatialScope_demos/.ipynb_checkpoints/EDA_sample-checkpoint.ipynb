# Load Visium and MERFISH Data

### Preprocessing Pipeline for Visium Data

The preprocessing pipeline for the scRNA-seq dataset (`ad_sc`) consists of the following key steps to prepare the data for downstream analysis:

**Pipeline Summary Table**

| **Step**                     | **Method/Function**                    | **Purpose**                                   |
|:-----------------------------|:--------------------------------------|:---------------------------------------------|
| Data Loading                 | `sc.read()`                           | Load processed scRNA-seq data                |
| Region Filtering             | `ad_sc.obs` filtering                 | Keep cells from specific brain region (VISp) |
| Count Scale Adjustment       | `np.exp(ad_sc.X) - 1` (if needed)     | Convert from log-transformed to raw counts   |
| Normalization                | `sc.pp.normalize_total()`             | Adjust for sequencing depth differences      |
| Gene Filtering               | `sc.pp.filter_genes()`                | Remove low-expressed and mitochondrial genes |
| Highly Variable Genes (HVGs) | `sc.pp.highly_variable_genes()`       | Identify informative, variable genes         |
| Marker Identification        | `sc.tl.rank_genes_groups()`           | Detect marker genes for cell types           |
| Ligand-Receptor Annotation   | External file integration             | Add genes for cell-cell interaction analysis |
| Log-Transformation           | `sc.pp.log1p()`                       | Stabilize variance in gene expression        |
| Dimensionality Reduction     | `sc.pp.pca()`                         | Reduce dimensionality with PCA               |
| Neighbors Graph              | `sc.pp.neighbors()`                   | Construct graph for clustering               |
| Visualization                | `sc.tl.umap()`                        | Project data into 2D for visualization       |

**Key Points**

- **Quality Control:** Removes low-quality cells and genes for cleaner analysis.
- **Normalization & Scaling:** Ensures data comparability across cells.
- **Dimensionality Reduction:** Simplifies complex data for easier visualization.
- **Marker Identification:** Highlights key genes driving cell-type differences.

### Preprocessing Pipeline for MERFISH Data

The preprocessing pipeline for the MERFISH dataset involves steps for data loading, quality control, and visualization.

**Pipeline Summary Table**

| **Step**                        | **Function/Method**                     | **Purpose**                                  |
|:--------------------------------|:---------------------------------------|:--------------------------------------------|
| Data Loading                    | `sc.read()`                            | Load MERFISH spatial and single-cell data    |
| Cell Type Annotation Check      | `value_counts()`                       | Validate distribution of annotated cell types |
| Dimensionality Reduction (UMAP) | `sc.pl.umap()`                         | Visualize clustering of different cell types |
| Color Mapping                   | `color_dict`                           | Assign distinct colors to each cell type     |
| Spatial Visualization           | `sns.scatterplot()`                    | Display spatial organization of cell types   |

**Key Points**

- **Quality Control:** Ensures balanced distribution of annotated cell types.
- **Dimensionality Reduction:** Uses UMAP to observe cell clustering patterns.
- **Spatial Mapping:** Visualizes how different cell types are distributed in tissue.

# Dual Graph Autoencoder (DGAE) Model
The goal is to integrate two different spatial transcriptomics datasets **Visium and MERFISH** into a shared latent space.

Model Architecture: 

- Two Encoders (GCN): One for Visium, one for MERFISH. Graph Convolutional Networks (GCNs) process graph data, capturing both gene expression and spatial relationships.

- Shared Latent Space: Aligns biological signals, ignoring technical noise. Combines outputs from both encoders.

- Decoders: Reconstruct input data.

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch_geometric.nn import GCNConv

class DualGraphAutoencoder(nn.Module):
    def __init__(self, input_dim, hidden_dim, latent_dim):
        super(DualGraphAutoencoder, self).__init__()

        # Encoders for Visium and MERFISH graphs
        self.encoder_visium = GCNConv(input_dim, hidden_dim)
        self.encoder_merfish = GCNConv(input_dim, hidden_dim)

        self.encoder_shared_visium = GCNConv(hidden_dim, latent_dim)
        self.encoder_shared_merfish = GCNConv(hidden_dim, latent_dim)

        # Decoders to reconstruct gene expression
        self.decoder_visium = nn.Linear(latent_dim, input_dim)
        self.decoder_merfish = nn.Linear(latent_dim, input_dim)

    def forward(self, visium_data, merfish_data):
        x_visium, edge_index_visium = visium_data['x'], visium_data['edge_index']
        x_merfish, edge_index_merfish = merfish_data['x'], merfish_data['edge_index']
    
        # Encode Visium
        h_visium = F.relu(self.encoder_visium(x_visium, edge_index_visium))
        z_visium = self.encoder_shared_visium(h_visium, edge_index_visium)
    
        # Encode MERFISH
        h_merfish = F.relu(self.encoder_merfish(x_merfish, edge_index_merfish))
        z_merfish = self.encoder_shared_merfish(h_merfish, edge_index_merfish)
    
        # Decode separately
        recon_visium = self.decoder_visium(z_visium)    # Match visium_features shape
        recon_merfish = self.decoder_merfish(z_merfish)  # Match merfish_features shape
    
        # Shared latent space for downstream tasks (optional)
        z_shared = torch.cat((z_visium, z_merfish), dim=0)
    
        return recon_visium, recon_merfish, z_shared

```

# Model Training

```python
# Hyperparameters
input_dim = visium_data.shape[1]  # Number of genes
hidden_dim = 128
latent_dim = 64
epochs = 100
learning_rate = 0.001

# Initialize Model, Loss, Optimizer
model = DualGraphAutoencoder(input_dim, hidden_dim, latent_dim)
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
loss_fn = nn.MSELoss()

# Convert data to tensors
visium_features = torch.tensor(visium_data.X.toarray() if hasattr(visium_data.X, 'toarray') else visium_data.X, dtype=torch.float32)
merfish_features = torch.tensor(merfish_data.X.toarray() if hasattr(merfish_data.X, 'toarray') else merfish_data.X, dtype=torch.float32)

# Placeholder for edge indices (update with real spatial graph)
edge_index_visium = torch.tensor([[0, 1], [1, 0]], dtype=torch.long)
edge_index_merfish = torch.tensor([[0, 1], [1, 0]], dtype=torch.long)

# Training Loop
for epoch in range(epochs):
    model.train()
    optimizer.zero_grad()

    # Forward Pass
    recon_visium, recon_merfish, z_shared = model(
        visium_data={'x': visium_features, 'edge_index': edge_index_visium},
        merfish_data={'x': merfish_features, 'edge_index': edge_index_merfish}
    )

    # Compute Loss
    loss_visium = loss_fn(recon_visium, visium_features)
    loss_merfish = loss_fn(recon_merfish, merfish_features)
    
    # Total Loss
    loss = loss_visium + loss_merfish

    # Backpropagation
    loss.backward()
    optimizer.step()

    if (epoch + 1) % 10 == 0:
        print(f'Epoch {epoch + 1}, Loss: {loss.item():.4f}')
```

# Latent Space Visualization: UMAP Visualization

```python
import umap
import matplotlib.pyplot as plt

# Reduce dimensionality
umap_model = umap.UMAP(n_neighbors=15, min_dist=0.1, metric='cosine')
embedding = umap_model.fit_transform(z_shared.detach().numpy())

# Plot
plt.figure(figsize=(8, 6))
plt.scatter(embedding[:, 0], embedding[:, 1], c='blue', s=5)
plt.title('Integrated Latent Space (Visium + MERFISH)')
plt.show()
```

# KMeans Clustering

```python
from sklearn.cluster import KMeans

# Apply KMeans to the latent space
n_clusters = 11
kmeans = KMeans(n_clusters=n_clusters, random_state=4)
clusters = kmeans.fit_predict(z_shared.detach().numpy())

# Add clusters to the original data
visium_data.obs['cluster'] = clusters[:visium_data.shape[0]]
merfish_data.obs['cluster'] = clusters[visium_data.shape[0]:]

# Visualize the clusters
plt.figure(figsize=(8, 6))
plt.scatter(embedding[:, 0], embedding[:, 1], c=clusters, cmap='tab10', s=5)
plt.title('KMeans Clustering of Integrated Data')
plt.colorbar(label='Cluster')
plt.show()
```

```python
import seaborn as sns

# Plotting MERFISH spatial distribution by clusters
plt.figure(figsize=(8, 6))
sns.scatterplot(
    x=merfish_data.obs['x'],
    y=merfish_data.obs['y'],
    hue=merfish_data.obs['cluster'],
    palette='tab10',
    s=20,
    legend='full'
)

plt.gca().invert_yaxis()
plt.title('Spatial Distribution of Clusters (MERFISH)')
plt.legend(title='Cluster', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# List all unique cell types
unique_cell_types = merfish_data.obs['cell_type'].unique()
# print("Known Cell Types in MERFISH Data:")
# print(unique_cell_types)

# Plotting MERFISH spatial distribution by known cell types using seaborn
plt.figure(figsize=(8, 6))
sns.scatterplot(
    x=merfish_data.obs['x'],
    y=merfish_data.obs['y'],
    hue=merfish_data.obs['cell_type'],
    palette='tab10',
    s=20,
    legend='full'
)

plt.gca().invert_yaxis()
plt.title("Spatial Distribution of Known Cell Types (MERFISH)")
plt.legend(title='Cell Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
